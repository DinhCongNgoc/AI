```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import Flatten, Dense,Dropout
```

```python
model = Sequential()
model.add(Conv2D(128,(3,3),input_shape=(256,256,3),activation = 'relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(32,(3,3),activation = 'relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(128,activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(10,activation ='softmax'))
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
```

```python
training_set=train_datagen.flow_from_directory('/content/gdrive/MyDrive/Foods/train',target_size=(256,256),batch_size=32,class_mode='categorical')
```

```python
test_set=train_datagen.flow_from_directory('/content/gdrive/MyDrive/Foods/test',target_size=(256,256),batch_size=32,class_mode='categorical')
```

```python
history=model.fit(training_set,epochs=100,validation_data=test_set,verbose=1)
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image
from keras import models
import matplotlib.pyplot as plt 
import matplotlib.image as img
model.save('Final.h5')
model_CNN=models.load_model('Final.h5')
test_image = image.load_img('/content/gdrive/MyDrive/Foods/validation/che/image 1.jpg')
plt.imshow(test_image)
img = image.img_to_array(test_image)
img = img.reshape(1,256,256,3)
img = img.astype('float32')
img = img/255
x = np.argmax(model_CNN.predict(img),axis=1)
print(x)
```
